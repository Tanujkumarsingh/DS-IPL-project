```python
# team bat first and win 
# Team batted first and won the match
batting_first_wins = ipl[ipl['result'] == "runs"]
```

```python
batting_first_wins.head()
```

```python
batting_first_counts = batting_first_wins['winner'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(batting_first_counts.index, batting_first_counts.values, color='skyblue')
plt.title('Teams Winning Matches Batting First (By Runs)')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)  # Proper way to rotate x-axis labels
plt.tight_layout()
plt.show()
```

```python
batting_first_wins['winner'].value_counts()
```

```python
plt.figure(figsize=(7, 7))
plt.bar(list(batting_first_wins['winner'].value_counts()[0:5].keys()), list(batting_first_wins['winner'].value_counts()[0:5]), color= ["green", "red","blue", "orange", "pink"])
plt.title("teams won")
plt.xlabel("teams")
plt.ylabel("no. of matches")
plt.xticks(rotation=45) 
plt.show()
```

```python
# Making a pie chart
plt.figure(figsize=(7, 7))
plt.pie(list(batting_first_wins['winner'].value_counts()), labels=list(batting_first_wins['winner'].value_counts().keys()), autopct = '%0.1f%%')
plt.show()
```
